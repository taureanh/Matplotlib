## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results, how='outer', on = 'Mouse ID' )

# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice.
miceid_count = combined_df["Mouse ID"].count()
miceid_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mice_duplicate = combined_df[combined_df.duplicated(['Mouse ID', 'Timepoint'])]

 
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
print(mice_duplicate)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
duplicate_drop = combined_df.drop_duplicates(subset=["Mouse ID", "Timepoint"], keep = False)
duplicate_drop 

```

```python
# Checking the number of mice in the clean DataFrame.
new_count = duplicate_drop["Mouse ID"].count()
new_count

```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.
count = duplicate_drop["Drug Regimen"].value_counts()
count
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function


summary = duplicate_drop.groupby(['Drug Regimen', 'Timepoint'])
mean = summary["Tumor Volume (mm3)"].mean()
dfmean = pd.DataFrame(mean)
dfmean

```

```python
median = summary["Tumor Volume (mm3)"].median()
dfmedian = pd.DataFrame(median)
dfmedian
```

```python
std_dev = summary["Tumor Volume (mm3)"].std()
dfstd_dev = pd.DataFrame(std_dev)
dfstd_dev
```

```python
sem = summary["Tumor Volume (mm3)"].sem()
dfsem = pd.DataFrame(sem)
dfsem
```
